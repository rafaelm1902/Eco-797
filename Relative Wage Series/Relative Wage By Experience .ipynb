```stata
egen temp1=sum(labor_supply_weight),by(year)
gen normalized_labor_supply_weight = labor_supply_weight/temp1
egen average_labor_weight=mean(normalized_labor_supply_weight),by(schooling female experience_cat) // average labor supply
drop temp1
```

```stata
 * High school By experience
egen m1=sum(predicted_log_wages*labor_supply_weight*(schooling==2)), by(year experience_cat)
egen m2=sum(labor_supply_weight*(schooling==2)), by(year experience_cat)
gen hs_wage_exp = m1/m2
drop m1 m2
```

```stata
 * By experience college-plus
egen m1=sum(predicted_log_wages*labor_supply_weight*(schooling==4 | schooling==5)), by(year experience_cat)
 egen m2=sum(labor_supply_weight*(schooling==4 | schooling == 5)), by(year experience_cat)
 gen college_wage_exp=m1/m2
 drop m1 m2
```

```stata
gen experience_college_hs_gap = college_wage_exp - hs_wage_exp
```

```stata
bysort year experience_cat: keep if _n==1
```
