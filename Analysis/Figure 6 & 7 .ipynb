```stata
cd
use "..\Data\experience_relative_wage"
sort year
merge m:1 year experience_cat using "../Data/experience_relative_supply", keep(match) nogenerate
replace year = year-1
merge m:1 year using "../Data/unemp_min", keep(match) nogenerate
sort year
merge m:1 year using "../Data/pce", keep(match) nogenerate
sort year
```

```stata
sum
```

```stata
gen expw09 = 100* (experience_college_hs_gap) if experience_cat == 1
gen expw29 = 100* (experience_college_hs_gap) if experience_cat == 3
```

```stata
twoway connected expw09 expw29 year if year <=2005, sort msize(small small) ///
msymbol(o s) mcolor(orange_red navy) lcolor(orange_red navy) xtitle(Earning Year) xlabel(1963(6)2005) ///
ytitle(Log Wage Gap x 100) ylabel(30(5)70, angle(0)) graphregion(color(white)) xsize(6) ysize(3.5) ///
legend(size(medium) label( 1 "Experience 0-9") ///
label( 2 "Experience 20-29") region(lstyle(none)))
graph export "..\Results\figure6a-05.pdf", as(pdf) replace
```

```stata
twoway connected expw09 expw29 year, sort msize(small small) ///
msymbol(o s) mcolor(orange_red navy) lcolor(orange_red navy) xtitle(Earning Year) xlabel(1963(7)2019) ///
ytitle(Log Wage Gap x 100) ylabel(30(5)75, angle(0)) graphregion(color(white)) xsize(6) ysize(3.5) ///
legend(size(medium) label( 1 "Experience 0-9") ///
label( 2 "Experience 20-29") region(lstyle(none))) ///
xline(2005, lwidth(thin) lcolor(cranberry) lpattern(dash)) 
graph export "..\Results\figure6a-19.pdf", as(pdf) replace
```
