```stata
summarize pce if inlist(year, 2000), meanonly //  2000 earning year, since we subtracte year - 1 
scalar pce2000=r(mean)
scalar list pce2000


gen real_min_wage = ((pce2000*nominal_minwage)/pce)
gen log_real_min_wage = log(real_min_wage)
```
